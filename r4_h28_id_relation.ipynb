```python
import glob
import os
import re

import pandas as pd

files = glob.glob('./raw/gsheets/*編集用/行き先がないID.csv')
removed_ids= pd.DataFrame([])
for file in files:
    if os.path.getsize(file)<8:
        continue
    df = pd.read_csv(file)
    dir = os.path.basename(os.path.dirname(file))
    tab = re.match(r'(.+?)編集用',dir).group(1)
    df["タブ名"]=tab
    removed_ids = pd.concat([removed_ids,df.loc[:,["タブ名","H28ID","理由・コメント"]]],axis=0)

removed_ids

```
