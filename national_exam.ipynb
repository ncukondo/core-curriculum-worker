```python
import csv
import pandas as pd
import jaconv

raw=pd.read_csv("./raw/医師国家試験出題基準の改定に向けた提言のための研究(門田・伴)-data.csv")
raw=raw[raw["番号"]!="通番"]
for name in raw.columns.values:
    raw[name]=raw[name].str.replace(r'\n|\r',"",regex=True)
raw=raw.applymap(lambda x:jaconv.z2h(str(x),kana=False, digit=True, ascii=True))
raw["番号"]=raw["番号"].str.zfill(4)
raw.to_csv("./raw/national_exam.csv",encoding="utf_8_sig",quoting=csv.QUOTE_NONNUMERIC,index=False)

raw
```

```python
to_delete=raw=raw[raw["削除"].astype(int)>0]
to_delete=to_delete.loc[:,["大項目","中項目","小項目","削除"]]
to_delete.to_csv("./dist/draft_cs/national_exam_to_delete.csv",encoding="utf_8_sig",quoting=csv.QUOTE_NONNUMERIC,index=False)
to_delete
```
