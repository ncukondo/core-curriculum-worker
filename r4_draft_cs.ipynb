```python
#!/usr/bin/env python
# coding: utf-8

import csv
import pandas as pd


r4_l234=pd.read_csv("./raw/draft/R4_CS_draft_L234.csv")
h28_working=pd.read_csv("./raw/draft/H28_CS_draft.csv")

h28_working
```

```python
r4_l4_from_h28=h28_working.loc[:,['第3層','第4層','H28対応項目','メモ']].dropna(subset=['第3層'])
r4_l4=pd.merge(r4_l234.loc[:,["第2層","第3層"]],r4_l4_from_h28,how="outer",on="第3層")
r4_l4.to_csv("./dist/draft_cs/l4.csv",encoding="utf_8_sig",index="False")

r4_l4
```

```python
departments=h28_working.loc[:,['診療科','診療科第3層','text4','H28対応項目','メモ']].dropna(subset=['診療科']).fillna("")
departments=departments\
    .groupby(['診療科','診療科第3層'],sort=False)\
    .apply(lambda x:pd.DataFrame({
        "H28対応項目":[",".join(list(x.H28対応項目))],
        "text4":[",".join(list(set(x.text4)))],
        "メモ":["\n".join(list(x.メモ))],
        }))\
    .reset_index()\
    .rename(columns={"診療科第3層":"第3層","text4":"診療科","診療科":"第4層"})\
    .loc[:,["第3層","第4層","H28対応項目","診療科"]]
departments.to_csv("./dist/draft_cs/departments.csv",encoding="utf_8_sig",quoting=csv.QUOTE_NONNUMERIC,index=False)
departments
```

```python
departments_per_name=h28_working.loc[:,['診療科','第4層','text4','H28対応項目','メモ']].dropna(subset=['診療科']).fillna("")
departments_per_name=departments_per_name\
    .groupby(['text4'],sort=False)\
    .apply(lambda x:pd.DataFrame({
        "H28対応項目":[",".join(list(x.H28対応項目))],
        "第4層":[",".join(list(x.第4層))],
        "メモ":[",".join(list(x.メモ))],
        }))\
    .reset_index()\
    .rename(columns={"text4":"診療科"})\
    .loc[:,["診療科","H28対応項目","第4層","メモ"]]
departments_per_name.to_csv("./dist/draft_cs/departments_per_name.csv",encoding="utf_8_sig",quoting=csv.QUOTE_NONNUMERIC,index=False)
departments_per_name
```

```python
symptoms=h28_working.loc[:,['症候','第4層','H28対応項目','メモ']].dropna(subset=['症候']).fillna("")
symptoms=symptoms.groupby(['症候'],sort=False)\
            .apply(lambda x:pd.DataFrame({
                "H28対応項目":[",".join(list(x.H28対応項目))],
                "第4層":["\n".join(list(x.第4層))],
                "メモ":["\n".join(list(x.メモ))],
                })).reset_index().drop("level_1",axis=1)
symptoms.to_csv("./dist/draft_cs/symptoms.csv",encoding="utf_8_sig",quoting=csv.QUOTE_NONNUMERIC,index=False)
symptoms
```

```python
symptoms_per_disease=h28_working.loc[:,['症候','text3','H28対応項目','第4層','メモ']].dropna(subset=['症候']).fillna("")
symptoms_per_disease=symptoms_per_disease.groupby(['text3'],sort=False)\
            .apply(lambda x:pd.DataFrame({
                "H28対応項目":[",".join(list(x.H28対応項目))],
                "第4層":[",".join(list(x.第4層))],
                "症候":["\n".join(list(x.症候))],
                "メモ":["\n".join(list(x.メモ))],
                })).reset_index().drop("level_1",axis=1)
symptoms_per_disease.to_csv("./dist/draft_cs/symptoms_per_disease.csv",encoding="utf_8_sig",quoting=csv.QUOTE_NONNUMERIC,index=False)
symptoms_per_disease
```
