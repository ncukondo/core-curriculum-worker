```python
import pandas as pd
import re

from lib.utils import DATA_DIR,TABLES_DIR,OUTPUT_DIR

with open(f"{DATA_DIR}/outcomes.md","r") as f:
    outcomes_md = f.read()
with open(f"{DATA_DIR}/outcomes_l1.md","r") as f:
    outcomes_md_l1 = f.read()



definitions = pd.read_csv(f"{DATA_DIR}/definitions.csv").set_index("データ中の記述")
def_keys = "|".join([re.escape(x) for x in definitions.index])
def_dict = {k: {
        "in_text":definitions.at[k,"文書中の記述"],
        "ref":definitions.at[k,"文書中の注釈"]
    } for k in definitions.index}
def_dict

def_list = []
i=0

def replace_fn(reg):
    global i
    global def_list
    s=reg.group(0)
    if s in def_dict:
        i +=1
        def_list.append(f'[^{i}]: {def_dict[s]["ref"]}')
        return f'{def_dict[s]["in_text"]}[^{i}]'
    else:
        return s

outcomes_md = re.sub(def_keys,replace_fn,outcomes_md)
outcomes_md = outcomes_md +"\n\n"+"\n".join(def_list)+"\n\n"
outcomes_md


outcomes_md=re.sub(r"(\n## )",r"\n\\newpage\n\1",outcomes_md[1:])


outcomes_md=r"""---
title: "医学教育モデル・コア・カリキュラム"
header-includes: 
- \usepackage{multirow}
- \usepackage{xltabular}
- \usepackage{longtable}
- \usepackage{ltablex}
- \usepackage {booktabs}
- \usepackage {ulem}
figureTitle: "図 "
tableTitle: "表 "
listingTitle: "コード "
figPrefix: "図."
eqnPrefix: "式."
tblPrefix: "表."
lstPrefix: "コード."
---

"""+outcomes_md

outcomes_md += f"\n\\newpage\n\n## 別表\n\n"

tables = ""

with open(f"{OUTPUT_DIR}/tables.tex","r") as f:
    latex_tables = f.read()

outcomes_md += latex_tables

outcomes_md = "\n\\newpage\n\n# 第2章 学修目標\n\n"+outcomes_md
outcomes_md_l1 = "\n\\newpage\n\n# 第1章 資質・能力\n\n"+outcomes_md_l1
outcomes_md = outcomes_md_l1 +"\n\n"+ outcomes_md


```

```python
with open(f"{OUTPUT_DIR}/outcomes_for_tex.md","w") as f:
    f.write(outcomes_md)

print("output... outcomes_for_tex.md")

outcomes_md
```
