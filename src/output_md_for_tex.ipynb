```python
import pandas as pd
import re

from lib.utils import DATA_DIR,TABLES_DIR,OUTPUT_DIR,TABLES_INDEX

with open(f"{OUTPUT_DIR}/outcomes.md","r") as f:
    outcomes_md = f.read()
with open(f"{OUTPUT_DIR}/outcomes_l1.md","r") as f:
    outcomes_md_l1 = f.read()

def replace_table_link(text:str)->str:
    table_index = pd.read_csv(f'{TABLES_INDEX}',encoding="utf_8_sig").fillna("").set_index("id")
    def id_to_table_caption(id:str):
        try:
            index =  table_index.at[id,"index"]
            return f"表{index}"
        except KeyError:
            return ""
    def replace_func(reg:re.match)->str:
        id = reg.group(1)
        whole = reg.group(0)
        caption = id_to_table_caption(id)
        if caption:
            return "\hyperref["+id+"]{"+caption+"}"
        else:
            return whole
    return re.sub(r"\[@tbl:([^\]]+)\]",replace_func,text)

outcomes_md = replace_table_link(outcomes_md)



outcomes_md=re.sub(r"(\n## )",r"\n\\newpage\n\1",outcomes_md[1:])


outcomes_md=r"""---
title: "医学教育モデル・コア・カリキュラム"
header-includes: 
- \usepackage{multirow}
- \usepackage{xltabular}
- \usepackage{longtable}
- \usepackage{ltablex}
- \usepackage {booktabs}
- \usepackage {ulem}
figureTitle: "図 "
tableTitle: "表 "
listingTitle: "コード "
figPrefix: "図."
eqnPrefix: "式."
tblPrefix: "表."
lstPrefix: "コード."
---

\captionsetup{labelformat=empty,labelsep=none}
"""+outcomes_md

outcomes_md += f"\n\\newpage\n\n## 別表\n\n"

tables = ""

with open(f"{OUTPUT_DIR}/tables.tex","r") as f:
    latex_tables = f.read()

outcomes_md += latex_tables

with open(f"{OUTPUT_DIR}/about_description.md","r") as f:
    about_description = f.read()

new_page= "\n\\newpage\n\n"
outcomes_md = new_page+"# 第2章 学修目標\n\n"+outcomes_md
outcomes_md_l1 = new_page+"# 第1章 資質・能力\n\n"+outcomes_md_l1
outcomes_md = new_page+about_description+new_page+outcomes_md_l1 +"\n\n"+ outcomes_md


```

```python
with open(f"{OUTPUT_DIR}/outcomes_for_tex.md","w") as f:
    f.write(outcomes_md)

print("output... outcomes_for_tex.md")

outcomes_md
```
